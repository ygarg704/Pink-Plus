# PINK+ (EXPERIMENTAL DATA VISUALIZATION)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
from pylab import savefig
from pywaffle import Waffle
```

```python
ds = pd.read_csv('BSC.csv')
ds.head()
```

```python
#Females who heard about BSE but still have not practiced it yet.

ds1 = ds.loc[ds['Have you heard of BSE (Breast Self-Examination)?'] == 'Yes']
ds1 = ds1.loc[ds['Gender'] == 'Female']
ds1 = ds1[['Gender', 'Have you heard of BSE (Breast Self-Examination)?', 'If yes, then have you ever practiced it?']]

fig = plt.figure(
    FigureClass=Waffle, 
    rows=6,
    figsize=(10, 8),
    values=ds1['If yes, then have you ever practiced it?'].value_counts(),
    colors=("#FF9393", "#cc4339"),
    labels=list(['No - 43.54 %', 'Yes - 56.45 %']),
    icons='child', icon_legend = True,
    title={'label': 'Females who heard about BSE but still have not praticed it yet', 'loc': 'center'}
)

fig.savefig("outputs/output_female.png", dpi = 400, transparent = True)
```

```python
#Males who heard about BSE but still have not practiced it yet.

ds1 = ds.loc[ds['Have you heard of BSE (Breast Self-Examination)?'] == 'Yes']
ds1 = ds1.loc[ds['Gender'] == 'Male']
ds1 = ds1[['Gender', 'Have you heard of BSE (Breast Self-Examination)?', 'If yes, then have you ever practiced it?']]

fig = plt.figure(
    FigureClass=Waffle, 
    rows=3,
    figsize=(10, 3),
    values=ds1['If yes, then have you ever practiced it?'].value_counts(),
    colors=["#8842FF"],
    labels=ds1['If yes, then have you ever practiced it?'].unique(),
    icons='child', icon_legend = True,
    title={'label': 'Males who heard about BSE but still have not praticed it yet', 'loc': 'center'}
)

fig.savefig("outputs/output_male.png", dpi = 400, transparent = True)
```

```python
#Percentage of men who knew that they could also be diagnosed with breast cancer.

ds_menc = ds.loc[ds['Gender'] == 'Male']

fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    figsize=(10, 6),
    values=ds_menc['Did you know that men can also be diagnosed with Breast Cancer?'].value_counts(),
    icons='child', icon_legend = True,
    colors=("#8842FF", "#cc4339"),
    labels=list(['No - 63.46 %', 'Yes - 36.53 %']),
    title={'label': 'Percentage of men who knew that they could also be diagnosed with breast cancer', 'loc': 'center'}
)

fig.savefig("outputs/output1_male.png", dpi = 400, transparent = True)
```

```python
#Percentage of women who knew that men could also be diagnosed with breast cancer.

ds_menc = ds.loc[ds['Gender'] == 'Female']

fig = plt.figure(
    FigureClass=Waffle,
    rows=6,
    figsize=(10, 7),
    values=ds_menc['Did you know that men can also be diagnosed with Breast Cancer?'].value_counts(),
    icons='child', icon_legend = True,
    colors=("#FF9393", "#cc4339"),
    labels=list(['Yes - 51 %', 'No - 49 %']),
    title={'label': 'Percentage of women who knew that men could also be diagnosed with breast cancer', 'loc': 'center'}
)

fig.savefig("outputs/output1_female.png", dpi = 400, transparent = True)
```

```python
#Graph 3
ds2 = ds[['Age', 'Which of the following can be the best way to spread awareness about Breast Cancer?']]
sns_plot = sns.catplot(x="Which of the following can be the best way to spread awareness about Breast Cancer?", hue="Age", data=ds2, kind="count", palette=['#FF9393','#8842FF','#d97e7e','#5528a1']);   
sns_plot.set_axis_labels('Which of the following can be the best way to spread awareness about Breast Cancer?', 'Number of people')
sns_plot.savefig("outputs/output2.png", dpi = 400, transparent = True)
```

```python
#Graph 4
ds2 = ds[['Gender', 'Which of the following can be the best way to spread awareness about Breast Cancer?']]
sns_plot = sns.catplot(x="Which of the following can be the best way to spread awareness about Breast Cancer?", hue="Gender", data=ds2, kind="count", palette=['#FF9393','#8842FF','#d97e7e','#5528a1']);   
sns_plot.set_axis_labels('Which of the following can be the best way to spread awareness about Breast Cancer?', 'Number of people')
sns_plot.savefig("outputs/output3.png", dpi = 400, transparent = True)
```
